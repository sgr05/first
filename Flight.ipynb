5)Identify any potential outliers in dep_delay column .

```python
data=df["dep_delay"]
q1=data.quantile(0.25)
q3=data.quantile(0.75)
IQR=q3-q1
lowBound=q1-1.5*IQR
uppBound=q3+1.5*IQR
outliers = data[(data < lowBound) | (data > uppBound)]
print(outliers)


```
